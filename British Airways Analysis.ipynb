```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('British Airways Summer Schedule Dataset -task1.xlsx.csv')
data.head()
```

```python
data.groupby(data.TIME_OF_DAY)[['FIRST_CLASS_SEATS','BUSINESS_CLASS_SEATS','ECONOMY_SEATS']].sum()
```

```python
time = data.groupby('TIME_OF_DAY')[['FIRST_CLASS_SEATS','TIER1_ELIGIBLE_PAX', 'BUSINESS_CLASS_SEATS',	'TIER2_ELIGIBLE_PAX',	'ECONOMY_SEATS','TIER3_ELIGIBLE_PAX']].sum()
time
```

```python
percentage = pd.DataFrame()
percentage['TIME_OF_DAY']=['Afternoon','Evening','Lunchtime','Morning']
percentage['ELIGIBLE_FIRST_CLASS_SEATS'] = (time.TIER1_ELIGIBLE_PAX/ time.FIRST_CLASS_SEATS * 100).values
percentage['ELIGIBLE_BUSINESS_CLASS_SEATS'] = (time.TIER2_ELIGIBLE_PAX/ time.BUSINESS_CLASS_SEATS * 100).values
percentage['ELIGIBLE_ECONOMY'] = (time.TIER3_ELIGIBLE_PAX / time.ECONOMY_SEATS * 100).values
```

```python
percentage
```

```python
destination = data.groupby('ARRIVAL_REGION')[['FIRST_CLASS_SEATS','TIER1_ELIGIBLE_PAX', 'BUSINESS_CLASS_SEATS',	'TIER2_ELIGIBLE_PAX',	'ECONOMY_SEATS','TIER3_ELIGIBLE_PAX']].sum()
destination
```

```python
destination_per = pd.DataFrame()
destination_per['Destination']=['Asia','Europe','Middle East','North America']
destination_per['ELIGIBLE_FIRST_CLASS_SEATS'] = (destination.TIER1_ELIGIBLE_PAX/ destination.FIRST_CLASS_SEATS * 100).values
destination_per['ELIGIBLE_BUSINESS_CLASS_SEATS'] = (destination.TIER2_ELIGIBLE_PAX/ destination.BUSINESS_CLASS_SEATS * 100).values
destination_per['ELIGIBLE_ECONOMY'] = (destination.TIER3_ELIGIBLE_PAX / destination.ECONOMY_SEATS * 100).values
```

```python
destination_per
```

```python
haul = data.groupby('HAUL')[['FIRST_CLASS_SEATS','TIER1_ELIGIBLE_PAX', 'BUSINESS_CLASS_SEATS',	'TIER2_ELIGIBLE_PAX',	'ECONOMY_SEATS','TIER3_ELIGIBLE_PAX']].sum()
haul
```

```python
percentage_haul = pd.DataFrame()
percentage_haul['HAUL'] = ['LONG','SHORT']
percentage_haul['ELIGIBLE_FIRST_CLASS_SEATS'] = (haul.TIER1_ELIGIBLE_PAX / haul.FIRST_CLASS_SEATS * 100).values
percentage_haul['ELIGIBLE_BUSINESS_CLASS_SEATS'] = (haul.TIER2_ELIGIBLE_PAX / haul.BUSINESS_CLASS_SEATS *100).values
percentage_haul['ELIGIBLE_ECONOMY'] = (haul.TIER3_ELIGIBLE_PAX / haul.ECONOMY_SEATS * 100).values
percentage_haul
```
